# Working on Dataset(1975-1999)

### i) Cleaning the data to seperate year,month,date,time,DST value and then storing them in form of a dictionary

```python
# Here re stands for a regular expression, regex(sometimes called a rational expression), which is 
#a sequence of characters that define a search pattern.This pattern is used by string searching algorithms
#for "find" or "find and replace" operations on strings.
import re
final_data={}
count1=1;
with open('1975-1999.txt') as file: 
    lines =file.readlines() #reading the file line by line
    for line in lines:
        l=line.split() #splitting the line on spaces
        if len(l)==0:
            continue #if line is empty then continue
        #Date contains Year,month and date part
        Date=(re.findall(r'([\d.]+)-([\d.]+)-([\d.]+)',l[0]))
        #time comtains hour,min and sec where hour and min is always 00:00 so they are not needed to be extracted
        Time=(re.findall(r'([\d.]+):([\d.]+):([\d.]+)',l[1]))
        Date=list(Date[0])
        Time=list(Time[0])
        Year=Date[0]
        Month=Date[1]
        date=Date[2]
        Hour=Time[0]
        #The third and final part of the list contains dayofyear and DST Value part
        Dayofyear=l[2]
        value=l[3]
        #Storing the values in the form of dictionary
        final_data[count1]=({'DayofYear':Dayofyear,'Date':date,'Month':Month,'Year':Year,'Time(Hrs)':Hour,'Dst Index':value})
        count1=count1+1
    file.close()
```

### ii) Converting Dictionary in DataFrame

```python
import pandas as pd
#pandas is an open source, library providing high-performance,easy-to-use data structures and data analysis tools for Python 
# DataFrame is tabular data structure with labeled axes (rows and columns).

df=(pd.DataFrame(final_data))#Converting a dictionary in a dataframe form

df=df.transpose()
df=(pd.DataFrame(df,columns=final_data[1].keys()))#Setting the column names
```

### iii) Displaying the Dataset 

```python
df
```

### iv) Saving the dataframe in form of csv file with name 'DST(1975-1999).csv'

```python
df.to_csv('DST(1975-1999).csv')
```

### v) Making respective data directories corresponding to the years

```python
#The OS module in Python provides a way of using operating system dependent functionality. 
import platform
import os
sep=''
if platform.system()=='Windows':#Select seperator as per os 
    sep='\\'
else:
    sep='/'
#sepecifies directory path where we want to keep the files
cwd=os.getcwd()
dirpath=cwd+sep+'YearData'
#if directory path already has the folder then we skip this step
if not os.path.exists(dirpath):
        os.makedirs(dirpath)
for i in range(1975,2000):
    name=dirpath+sep+'DST_'+str(i)+'.csv'
    df[(df.Year ==str(i))].to_csv(name)#selecting that particular year from the data and converting to csv file format
```

### vi) Making respective data directories corresponding to the months

```python
import os
dirpath1=cwd+sep+'MonthData'
if not os.path.exists(dirpath1):
        os.makedirs(dirpath1)
for i in range(1975,2000):
    dirpath2=cwd+sep+'MonthData'+sep+str(i)
    if not os.path.exists(dirpath2):
        os.makedirs(dirpath2)
    for j in range(1,13):
        mo=''
        if j<=9:
            mo='0'+str(j) #0-9 data is padded with a 0 at the beginning such that '1' is represented as '01'
        else:
            mo=str(j)
        #print(str(i)+" "+mo)
        name=dirpath2+sep+'DST_'+str(i)+'_'+mo+'.csv'
        (df[(df['Year']== str(i)) & (df['Month']==mo)]).to_csv(name)#selecting that particular year and month from the data and converting to csv file format
```
